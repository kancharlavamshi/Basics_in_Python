List Comprehension

List comprehension offers a shorter syntax when you want to create a new list based on the values of an existing list.

```python
countries = ["india", "srilanka", "china", "nepal", "bangladesh"]
newlist = []

for x in countries:
  if "e" in x:
    newlist.append(x)

print(newlist)
```

```python
newlist = []

for x in countries:
  if "a" in x:
    newlist.append(x)

print(newlist)
```

The Syntax

newlist = [expression for item in iterable if condition == True]

```python
## With list comprehension you can do all that with only one line of code

newlist = [x for x in countries if "a" in x]

print(newlist)
```

>Sort List Alphanumerically

List objects have a sort() method that will sort the list alphanumerically, ascending, by default

By default the sort() method is case sensitive, resulting in all capital letters being sorted before lower case letters
>>So if you want a case-insensitive sort function, use str.lower as a key function

>thislist.sort(key = str.lower)

>print(thislist)

```python
countries.sort()
print(countries)
```

```python
## Sort the list descending:
countries.sort(reverse = True)
print(countries)
```

**Set**

>A set is a collection which is both unordered and unindexed.

```python
## A set can contain different data types:

set1 = {"vam", 4, True, 21, "male"}
print(type(set1))
```

```python
## It is also possible to use the set() constructor to make a set. 
set2=set(("india", "srilanka", "china"))
print(set2)
```

You cannot access items in a set by referring to an index or a key.

But you can loop through the set items using a for loop, or ask if a specified value is present in a set, by using the in keyword.

```python

print("india" in set2)
```

**ADD Items**
Once a set is created, you cannot change its items, but you can add new items.

```python
set2.add("japan")

print(set2)
```

```python
## To add items from another set into the current set, use the update() method.
cities = {"Texas", "DC", "Dallas"}

set2.update(cities)

print(set2)
```

```python
## The object in the update() method does not have be a set, it can be any iterable object (tuples, lists, dictionaries etc.).

cities = ["Texas", "DC", "Dallas"]

set2.update(cities)

print(set2)
```

```python
set2
```

```python
## To remove an item in a set, use the remove(), or the discard() method.

set2.remove("Texas")

print(set2)
```

```python
## using discard method
set2.discard("india")

print(set2)
```

```python
for x in set2:
  print(x)
```

You can use the union() method that returns a new set containing all items from both sets, or the update() method that inserts all the items from one set into another

```python
set3 = set1.union(set2)
print(set3)
```

```python
set1.update(set2)
print(set1)
```

**Dictionaries**

>**Dictionaries** are used to store data values in key:value pairs.

>A dictionary is a collection which is ordered*, changeable and does not allow duplicates.

```python
dic = {
  "Student": "KK",
  "CLASS": "first",
  "year": 2014
}
print(dic)
```

```python
print(dic['year'])
```

```python
### There is also a method called get() that will give you the same result:
x = dic.get("year")
print(x)
```

```python
##As of Python version 3.7, dictionaries are ordered. In Python 3.6 and earlier, dictionaries are unordered.
```

When we say that dictionaries are ordered, it means that the items have a defined order, and that order will not change.

Unordered means that the items does not have a defined order, you cannot refer to an item by using an index

```python
##  Duplicates Not Allowed
##  Dictionaries cannot have two items with the same key:

dic = {
  "Student": "KK",
  "CLASS": "first",
  "year": 2014,
  "year": 2021,
}
print(dic),print(dic['year'])
```

```python
##  The keys() method will return a list of all the keys in the dictionary.

x = dic.keys()
print(x)
```

```python
dic = {
  "Student": "KK",
  "CLASS": "first",
  "year": 2014,
  "year": 2021,
}

x = dic.keys()

print(x) #before the change

dic["gender"] = "male"

print(x) #after the change
```

```python
x = dic.values()
print(x)
```

```python
##   The values() method will return a list of all the values in the dictionary
```

```python
dic = {
  "Student": "KK",
  "CLASS": "first",
  "year": 2014
}

x = dic.values()

print(x) #before the change

dic["year"] = 2021

print(x) #after the change
```

```python
## Change Values
##       You can change the value of a specific item by referring to its key name:

dic = {
  "Student": "KK",
  "CLASS": "first",
  "year": 2014
}
dic["year"] = 1990
print(dic)

###  The update() method will update the dictionary with the items from the given argument.

dic.update({"year": 1789})
print(dic)
```

```python
##  The pop() method removes the item with the specified key name
dic.pop("year")
print(dic)
#del dic["model"]
#print(dic)
```

```python
## loop through Dictionary
dic = {
  "Student": "KK",
  "CLASS": "first",
  "year": 2014
}
for x in dic:
  print(dic[x])
```

```python
for x in dic.values():
  print(x)
```

```python
for x in dic.keys():
  print(x)
```

You cannot copy a dictionary simply by typing dict2 = dict1, because: dict2 will only be a reference to dict1, and changes made in dict1 will automatically also be made in dict2.

There are ways to make a copy, one way is to use the built-in Dictionary method copy()

```python
mydict = dic.copy()
print(mydict)
```

```python
###Make a copy of a dictionary with the dict() function:


mydict = dict(dic)
print(mydict)
```

Nested Dictionaries

A dictionary can contain dictionaries, this is called nested dictionaries.

```python
country = {
  "TG" : {
    "name" : "kv",
    "year" : 2014
  },
  "TM" : {
    "name" : "vm",
    "year" : 2017
  },
  "AP" : {
    "name" : "kk",
    "year" : 2020
  }
}
print(country)
```
